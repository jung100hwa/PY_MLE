### 랭체인에 등록되어 있는 pdf로드 모듈을 테스트
1. PyPDFLoader 일반적
2. PyMuPDFLoader 속도가 엄청 빠름
3. PDFPlumberLoader 느림

스캔 내용은 약간씩 다름.
랭체인을 이용하면 저것들을 이용해 청크를 의미적으로 분리할 수 있음
그래서 랭체인을 쓰겠지

```python
import urllib.request
# from langchain.document_loaders import PyPDFLoader  # 구버전
from langchain_community.document_loaders import PyPDFLoader
from langchain_community.document_loaders import PyMuPDFLoader
# from langchain.document_loaders import PDFPlumberLoader
from langchain_community.document_loaders import PDFPlumberLoader
```

### PyPDFLoader 테스트

```python
%%time
loader = PyPDFLoader('2023_북한인권보고서.pdf')
pages = loader.load_and_split()
print("청크의 수: ", len(pages))

# 여기서 북한인권보고서 요약본을 사용하면 청크수가 0이 나옴. 맥락이 같아서 그런 것 같음
# %%time 앞에 아무것도 없어야 됨. 그래서 주석을 밑에 달고 있음
```

```python
pages[3]
```

```python
print(pages[3].page_content)
```

### PyMuPDFLoader 테스트

```python
%%time
loader = PyMuPDFLoader('2023_북한인권보고서.pdf')
pages = loader.load_and_split()
print('청크의 수 :', len(pages))

# 속도가 엄청 빠르네.
```

```python
pages[3]
```

### PDFPlumberLoader 테스트

```python
%%time
loader = PDFPlumberLoader('2023_북한인권보고서.pdf')
pages = loader.load_and_split()
print('청크의 수 :', len(pages))
```
