```python
import os
os.chdir(r"c:\projects\PY_MLE")

import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
sentences = ['nice great best amazing', 'stop lies', 'pitiful nerd', 'excellent work', 'supreme quality', 'bad', 'highly respectable']
y_train = [1, 0, 0, 1, 1, 0, 1]
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
vocab_size = len(tokenizer.word_index) + 1
print("vocab_size:{}".format(vocab_size))
```

```python
x_encoded = tokenizer.texts_to_sequences(sentences)
x_encoded
```

```python
max_len = max(len(word) for word in x_encoded)
print("max_len:{}".format(max_len))
```

```python
x_train = pad_sequences(x_encoded, maxlen=max_len)
y_train = np.array(y_train)
print(x_train)
print(y_train)
```

```python
# 파일을 다운로드 받아서 불어왔더 치고. 시간이 너무 많이 걸림
embedding_dict = dict()
f = open('glove.6B.50d.txt', encoding='utf8') # 차원이 100은 파일이 깨졌나 보네. 잘 안열리네

for line in f:
    word_vector = line.split()
    word = word_vector[0] # 구조가 이렇게 되어 있다. 맨앞이 단어이다

    word_vector_arr = np.array(word_vector[1:], dtype='float32')
    embedding_dict[word] = word_vector_arr
f.close()
```

```python
print("embedding vector: {}".format(len(embedding_dict)))
```

```python
print(embedding_dict['respectable'])
print('벡터의 차원수:{}'.format(len(embedding_dict['respectable'])))
```

```python
embedding_matrix = np.zeros((vocab_size, 50))
print(embedding_matrix)
print("embedding_matrix shape:{}".format(embedding_matrix.shape))
```

```python
print(tokenizer.word_index.items())
```

```python
print('great mapping:{}'.format(tokenizer.word_index['great']))
print(embedding_dict['great'])
```

```python
# 사전임베딩된 백터를 넣어준다.
for word, index in tokenizer.word_index.items():

    vector_value = embedding_dict.get(word)
    if vector_value is not None:
        embedding_matrix[index] = vector_value
```

```python
embedding_matrix[0]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, Flatten

output_dim = 50

model = Sequential(
    [
        Embedding(vocab_size, output_dim=output_dim, weights=[embedding_matrix], input_length=max_len, trainable=False),
        Flatten(),
        Dense(1, activation='sigmoid')
    ]
)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
model.fit(x_train, y_train, epochs=100, verbose=2)
```

```python
# 최종가중치 구하기
embedding_matrix_weight = model.layers[0].get_weights()
print(len(embedding_matrix_weight[0]))
print(embedding_matrix_weight)
```
