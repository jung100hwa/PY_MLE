청킹 테스트
청킹은 프로프트 입력 텍스트가 너무 길어서 특정 조건에 맞게 텍스트를 자르는 것

```python
import urllib.request
from langchain_text_splitters import RecursiveCharacterTextSplitter

urllib.request.urlretrieve("https://raw.githubusercontent.com/lovit/soynlp/master/tutorials/2016-10-20.txt", filename="2016-10-20.txt")

with open("2016-10-20.txt", encoding="utf-8") as f: file = f.read()
print('텍스트의 길이 :', len(file))
```

RecursiveCharacterTextSplitter() 를 이용하여 텍스트를 분할하는 text_splitter 객체를 만듭니다. 이때 chunk_size 의 값을 500 으로 지정하면 앞으로 text_splitter 로 텍스트를 분할할 때 각 분할된 청크는 길이 가 500 을 결코 넘지 않습니다. chunk_overlap 은 텍스트를 분할할 때 각 청크가 내용을 얼만큼 겹치게 할 것인지를 정하는 값으로 0 을 지정하면 각 청크의 내용이 겹치지 않습니다.

```python
# 여기서 분할하는게 아니고 spliter 도구만 생성
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=0)
```

이렇게 선언된 text_splitter 로 파이썬 문자열을 분할할 때는 create_documents() 를 사용합니다.

```python
texts = text_splitter.create_documents([file])
print('분할된 청크의 수:', len(texts))
```

```python
# 원문을 보려면
print(texts[1].page_content)
```

```python
# 각 청크의 길이를 확인, 청크 사이즈를 500으로 주었기 때문에 500이하이다.
# 딱 500이 안되는 이유는 500안에서 \n\n, \n, 공백 이렇게 자르기 때문이다.

print("1 chunk:", len(texts[1].page_content))
print("2 chunk:", len(texts[2].page_content))
```
