```python
urllib.request.urlretrieve("https://raw.githubusercontent.com/ukairia777/tensorflow-nlp-tutorial/main/09.%20Word%20Embedding/dataset/ted_en-20160408.xml", filename="ted_en-20160408.xml")
```

```python
targetXML = open('ted_en-20160408.xml', 'r', encoding='utf8')
target_text = etree.parse(targetXML)

parse_text = '\n'.join(target_text.xpath('//content/text()'))
parse_text
```

```python
content_txt = re.sub(r'\([^)]*\)', '', parse_text)
```

```python
# 문장토큰화
sent_text = sent_tokenize(content_txt)
sent_text
```

```python
normalized_text = []
for string in sent_text:
    string = re.sub(r'[^a-z0-9]+',' ',string.lower())
    normalized_text.append(string)

normalized_text
```

```python
# 단어토큰화
result = [word_tokenize(sentence) for sentence in normalized_text]
print("sample count:{}".format(len(result)))
```

```python
print(result[:5])
```

```python
# 훈련시키기
from gensim.models import Word2Vec, KeyedVectors
```

```python
# 아래 코드 자체가 학습. fit이 없나보네
model = Word2Vec(sentences=result, vector_size=100, window=5, min_count=5, workers=4, sg =0)
```

```python
# 테스트. 유사도, 연관도을 가진 임베딩 벡터를 만들었기 때문에 테스트 해보자
model_result = model.wv.most_similar('man')
model_result
```

```python
# 모델 저장
model.wv.save_word2vec_format('eng_w2v')
load_model = KeyedVectors.load_word2vec_format('eng_w2v')
load_model.most_similar('woman') # 모델로 리턴 받은 것 같은데 사용방법이 조금 다르네. 이걸 다 암기해야 되나???...
```
