```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.style.use('seaborn-v0_8-white')
```

기본직선을 표시

```python
fig = plt.figure()
ax = plt.axes()
plt.plot([0, 0.2, 0.4, 0.6, 0.8, 1] * 5)
plt.show()
```

곡선을 표시

```python
x = np.arange(0,10, 0.01)
fig = plt.figure()
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

라인 스타일을 지정

```python
plt.plot(np.random.randn(50).cumsum(), linestyle='-')
plt.plot(np.random.randn(50).cumsum(), linestyle='--')
plt.plot(np.random.randn(50).cumsum(), linestyle='-.')
plt.plot(np.random.randn(50).cumsum(), linestyle=':')
plt.show()
```

컬러스타일을 지정
이런게 #1234 이건 파워포인트로 정말 편한데

```python
plt.plot(np.random.randn(50).cumsum(), color = 'g')
plt.plot(np.random.randn(50).cumsum(), color = '#156082') # 이런것도 가능
plt.plot(np.random.randn(50).cumsum(), color='b')
plt.plot(np.random.randn(50).cumsum(), color='r')
plt.show()
```

라인과 컬러를 한번에

```python
plt.plot(np.random.randn(50).cumsum(), 'g-')
plt.plot(np.random.randn(50).cumsum(), 'b--')
plt.plot(np.random.randn(50).cumsum(), 'r-.')
plt.plot(np.random.randn(50).cumsum(), 'y:')
plt.show()
```

축 경계를 지정함

```python
plt.plot(np.random.randn(50))
plt.xlim(-1, 50)
plt.ylim(-5,5)
plt.show()
```

축 경계를 꽉차게

```python
plt.plot(np.random.randn(50))
plt.axis('tight')
plt.title("TITLE")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

레전드 표시

```python
plt.plot(np.random.randn(50), label='first')
plt.plot(np.random.randn(50), label='second')
plt.plot(np.random.randn(50), label='third')
plt.axis('tight')
plt.title("TITLE")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```
