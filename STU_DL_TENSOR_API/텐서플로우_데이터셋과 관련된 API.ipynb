tf.data에는 텐서플로우 데이터를 다룰수 있는 많은 함수들이 포함되어 있음
우리가 생성한 모델에 데이터를 넣기 위해서는 tf.data.Dataset 형태로 변환해야 한다.

from_tensors은 하나의 텐서를 생성한다.

```python
a = tf.data.Dataset.from_tensors(tf.constant([0,1,2,3,4]))
print(a)
```

from_tensor_slices 얘가 유용할 듯 여러개의 텐서를 한꺼번에 만들어 낸다.

```python
a = tf.data.Dataset.from_tensor_slices([10,20,30,40,50,60])
b = tf.data.Dataset.from_tensor_slices(tf.random.uniform([4,2]))
print(a)
print(b)
for item in a:
    print(item)
for item in b:
    print(item)
```

batch, repeat
batch: 자료를 끊는다. drop_remainder=True이면 나머지는 버린다. 개수가 안되니까
repeat: 텐서를 반복한다.
아래와 같은 예제에서 1~10까지 텐서를 3번 반복하면 1,...10,1....10,1...10인데 이것을 7단위로 끊는다.

```python
a = tf.data.Dataset.from_tensor_slices(np.arange(10))
b = a.repeat(3).batch(7)
for item in b:
    print(item)

print()

a = tf.data.Dataset.from_tensor_slices(np.arange(10))
b = a.repeat(3).batch(7, drop_remainder=True) 
for item in b:
    print(item)
```

map. filter
기존 파이썬과 동일 함
또는 원하는 형태로 데이터셋 형태로 가져 올 수 있다.

```python
dataset = Dataset.range(1,6)
dataset = dataset.map(lambda x: x * 2)
list(dataset.as_numpy_iterator())
```

```python
elements = [(1,'one'),(2,'two'),(3,'three')]
dataset = Dataset.from_generator(lambda: elements, (tf.int32, tf.string))

result = dataset.map(lambda x_int, y_str: x_int)
list(result.as_numpy_iterator())
```

shuffle, take
섞고, 가져오기

```python
data, info = tfds.load('imdb_reviews', with_info=True)
train_data = data['train']
train_data = train_data.batch(5).shuffle(5).take(2) # shuffle(5)에서 5는 그냥 버퍼사이즈
for data in train_data:
    print(data)
```

get_next 다음텐서의 값을 순차적으로 가져오기

```python
dataset = Dataset.range(2)
iterater = iter(dataset)
print(dataset)
print(iterater.get_next())
print(iterater.get_next())
```

```python
dataset = Dataset.from_tensor_slices(np.arange(1,10).reshape(3,3))
it = iter(dataset)

it.get_next()
it.get_next()
it.get_next()

```
