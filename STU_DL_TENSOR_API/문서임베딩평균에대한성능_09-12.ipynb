```python
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
os.chdir(r"c:\projects\PY_MLE")

import pandas as pd
import numpy as np
import re

from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential, load_model, Model
from tensorflow.keras.layers import Input, Dense, GlobalAveragePooling1D, Embedding
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
vocab_size = 20000

(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=vocab_size)
print("train data:{}".format(len(x_train)))
print("test data:{}".format(len(x_test)))
```

```python
# 이미 정수인코딩이 되어 있음
print(x_train[0])
print(y_train[0])
```

```python
x_len_train_mean = np.mean(list(map(len,x_train)), dtype=int)
x_len_test_mean = np.mean(list(map(len, x_test)), dtype=int)
print("x_train_len:{}".format(x_len_train_mean))
print("x_test_len:{}".format(x_len_test_mean))
```

```python
max_len = 400
x_train = pad_sequences(x_train, maxlen=max_len)
x_test = pad_sequences(x_test, maxlen=max_len)
```

```python
embedding_dim = 64

model = Sequential(
    [
        Embedding(vocab_size, embedding_dim),
        GlobalAveragePooling1D(),
        Dense(1, activation='sigmoid')
    ]
)
check_point_md = ModelCheckpoint('embedding_average_model.h5', monitor='val_acc', save_best_only=True, mode='max', verbose=1)
check_point_st = EarlyStopping(monitor='val_loss', patience=4, mode='min', verbose=1)

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
model.fit(x_train, y_train, batch_size=32, epochs=10, callbacks=[check_point_md, check_point_st], validation_split=0.2)
```

```python
loaded_model = load_model('embedding_average_model.h5')
print("정확도:{}".format(loaded_model.evaluate(x_test, y_test)[1]))
print(loaded_model.evaluate(x_test, y_test))
```
